### Import neccessary lib.

```python
import pandas as pd
import numpy as np
import math
from time import sleep
```

### get stock file

```python
df = pd.read_csv('top_50_stocks_for_buy.csv')
df = df.drop(df.columns[[0,1]], axis=1)
df.head()
```


# Buy code

### calculate buying price

```python
percentage = input('Enter the buying percentage: ')
if not percentage.isdigit():
    print('Please enter the numeric value: ')
    percentage = input('Enter the buying percentage: ')

buying_percentage = float(percentage)
lst_buying_price = []
Ticker_list = []

for row in df.index:
    stock_price = df['Stock Price'][row]
    

    percent_price=(buying_percentage*stock_price)/100
    percent_price = float(int(percent_price * 1000))
    percent_price/=1000
#     print(percent_price)

    buying_price = stock_price + percent_price
    buying_price = float(int(buying_price * 1000))
    buying_price/=1000
    lst_buying_price.append(buying_price)
#     print(buying_price)  

    Ticker_list.append(df['Ticker'][row])
print(Ticker_list) 
```

### add buying price column in df

```python
df['Buying Price'] = lst_buying_price
df.head()
```

```python
want_to_buy = list(zip(Ticker_list,lst_buying_price,current_price))
print(want_to_buy)
```

```python
my_columns = ['Ticker','Stock Price', '5 Years Price Return','Number of shares to buy', 'Current_Price']


brought_df = pd.DataFrame(columns = my_columns)
left_df = pd.DataFrame(columns = my_columns)

# while temp_lst == []:
for Ticker1, buy_price, current in want_to_buy:
    
    if buy_price<current:
        brought_df = brought_df.append(df.loc[df['Ticker'] == Ticker1])
        
    else:
        left_df = left_df.append(df.loc[df['Ticker'] == Ticker1])
```
